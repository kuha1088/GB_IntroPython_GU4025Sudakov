```python
import pandas as pd
import seaborn as sns
```

```python
penguins = sns.load_dataset("penguins")
```

```python
penguins.head()
```

```python
penguins.describe()
```

```python
penguins.shape
```

```python
penguins.isnull().sum()
```

```python
new_penguins = penguins.dropna()
```

```python
sns.scatterplot(data = penguins, x = 'bill_length_mm', y = 'bill_depth_mm', hue='species')
```

```python
sns.scatterplot(data = penguins, x = 'body_mass_g', y = 'flipper_length_mm', hue='sex')
```

```python
sns.histplot(data = penguins, x = 'island')
```

```python
sns.histplot(data = penguins, x = 'island', y='body_mass_g')
```

```python
sns.histplot(data = penguins, x = 'flipper_length_mm')
```

```python
sns.histplot(new_penguins, x='island', y='body_mass_g', hue='sex')
```

```python
sns.heatmap(data=new_penguins.corr())
```
