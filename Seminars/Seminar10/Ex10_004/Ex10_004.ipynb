*Задача №69:*  
*1. Изобразить гистограмму по flipper_length_mm с оттенком height_group.*  
*Сделать анализ.*

```python
import pandas as pd
import seaborn as sns
```

```python
penguins = sns.load_dataset("penguins")
```

```python
penguins.head()
```

```python
new_penguins = penguins.dropna()
```

```python
new_penguins.loc[(new_penguins['bill_length_mm'] >= 42), 'bill_length_mm_new'] = 'high'
new_penguins.loc[(new_penguins['bill_length_mm'] > 35) & (new_penguins['bill_length_mm'] < 42), 'bill_length_mm_new'] = 'middle'
new_penguins.loc[(new_penguins['bill_length_mm'] <= 35), 'bill_length_mm_new'] = 'low'
```

```python
new_penguins.rename({'bill_length_mm_new': 'height_group'}, axis=1, inplace=True)
```

```python
new_penguins
```

```python
sns.histplot(data=new_penguins, x='flipper_length_mm', hue='height_group')
```
